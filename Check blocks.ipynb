```python
from patch25d import *
from matplotlib import pyplot as plt
import numpy as np

db = Patch25DB()
db.connect(h5file="test.small.h5")
x = db._db['test.small.tif']['data']['x']
y = db._db['test.small.tif']['data']['dist']
b = db._db['test.small.tif']['data']['label']

nsample = x.shape[0]
sample_idx = np.random.randint(0, nsample, (12,))
sample_idx.sort()

for i in sample_idx:
    xx = x[i,:,:,:,:,:]
    yy = y[i,:]
    bb = b[i,:]
    
    f, axarr = plt.subplots(1,3)
    axarr[0].imshow(np.squeeze(xx[0, 2, :, :, 0]))
    axarr[1].imshow(np.squeeze(xx[0, 2, :, :, 1]))
    axarr[2].imshow(np.squeeze(xx[0, 2, :, :, 2]))
    plt.title('dist=%.4f   label=%d' % (yy, bb))
plt.show()
```
